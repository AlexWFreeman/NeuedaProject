```python
import pandas as pd
df = pd.read_csv("./bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv")
df.head()
```

```python
df = df[~df['Open'].isna()]
df.head()
```

```python
import datetime
new_order = [col for col in df.columns]
df['Date'] = pd.to_datetime(df['Timestamp'], unit='s')
new_order.insert(1, 'Date')
df = df[new_order]
df.to_csv('./clean_data.csv', index=None, encoding="utf_8_sig")
df.head()
```

```python
df_day = pd.DataFrame()
df_group = df.groupby(df['Date'].dt.strftime('%Y-%m-%d'))
df_day['Open'] = df_group['Open'].apply(lambda i:i.iloc[0])
df_day['High'] = df_group['High'].max()
df_day['Low'] = df_group['Low'].min()
df_day['Close'] = df_group['Close'].apply(lambda i:i.iloc[-1])
df_day[['Volume_(BTC)','Volume_(Currency)','Weighted_Price']] = df_group[['Volume_(BTC)','Volume_(Currency)','Weighted_Price']].mean()
df_day = df_day.reset_index()
df_day['Timestamp'] = pd.to_datetime(df_day['Date']).apply(lambda x: x.timestamp()).astype(int)
df_day = df_day[new_order]
df_day.to_csv('./clean_data_by_day.csv', index=None, encoding="utf_8_sig")
df_day.head()
```
